```python
import pandas as pd
import plotly.express as px
import folium
from folium import plugins
from folium.plugins import MarkerCluster
folium.plugins.MarkerCluster()
from folium.plugins import MarkerCluster
```

```python
df = pd.read_excel('location_rating_with_lat_long_for_BNB.xlsx')
```

```python
df.head(10)
```

```python
def color_decider(rating):
    col = 'green' if rating == 10.0 else 'red'
    return col

def create_city_rating_map(df, city):
    m = folium.Map(location=[12.9716, 77.5946])
    df_plot = df.query(f"city_name == '{city}'")
    for index, row in df_plot.iterrows(): 
        col = color_decider(row['ugc_rating__location__review_score']) 
        folium.Marker(
        location = [row['geo__latitude'], row['geo__longitude']],
        popup = row['hotel_id'],
        icon = folium.Icon(color=col)).add_to(m)
    return m
```

```python
create_city_rating_map(df, 'Bengaluru')
```

```python
def color_decider(rating):
    col = 'green' if rating == 10.0 else 'crimson'
    return col

def create_city_rating_map_circle(df, city):
    m = folium.Map(location=[12.9716, 77.5946])
    df_plot = df.query(f"city_name == '{city}'")
    for index, row in df_plot.iterrows(): 
        col = color_decider(row['ugc_rating__location__review_score']) 
        folium.CircleMarker(
        location = [row['geo__latitude'], row['geo__longitude']],
        popup = row['hotel_id'],
        color = col,
        fill = False,
        fill_color = col,
        icon = folium.Icon(color=col)).add_to(m)
    return m
```

```python
create_city_rating_map_circle(df, 'Bengaluru')
```

```python
def color_decider(rating):
    col = 'green' if rating == 10.0 else 'red'
    return col

def create_city_rating_map_clustered(df, city_list):
    m = folium.Map(location=[20.5937,  78.9629], zoom_start=5)
    for city in city_list:
        marker_cluster = MarkerCluster().add_to(m)
        df_plot = df.query(f"city_name == '{city}'")
        for index, row in df_plot.iterrows(): 
            col = color_decider(row['ugc_rating__location__review_score']) 
            folium.Marker(
            location = [row['geo__latitude'], row['geo__longitude']],
            popup = row['hotel_id'],
            icon = folium.Icon(color=col)).add_to(marker_cluster)
    return m
```

```python
create_city_rating_map_clustered(df, ['Bengaluru', 'Shimla'])
```
